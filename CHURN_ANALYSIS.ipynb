# in below function we converted 0 and 1 value of SeniorCitizen to yes/no to make it easy

```python
def convert(value):
    if value==1:
        return "Yes"
    else:
        return "No"
dataframe['SeniorCitizen']=dataframe['SeniorCitizen'].apply(convert)
```

```python
#ax=sns.countplot(x='SeniorCitizen',data=dataframe ,hue= 'Churn')
#plt.show()


import seaborn as sns
import matplotlib.pyplot as plt

ax = sns.countplot(x='SeniorCitizen', data=dataframe, hue='Churn', palette=['lightgreen', 'salmon'])
plt.title("Churn by Senior Citizen Status")
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Number of Customers")
plt.legend(title='Churn')
plt.show()
```

```python
# Step 1: Get counts of Churn grouped by SeniorCitizen
grouped = dataframe.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Step 2: Convert counts to percentages
percent = grouped.div(grouped.sum(axis=1), axis=0) * 100

# Step 3: Plot stacked bar chart
ax = percent.plot(kind='bar', stacked=True, color=['lightgreen', 'salmon'])

# Step 4: Add percentage labels
for i, total in enumerate(percent.values):
    bottom = 0
    for j, value in enumerate(total):
        plt.text(i, bottom + value/2, f'{value:.1f}%', ha='center', va='center', fontsize=10)
        bottom += value

# Final touches
plt.title('Senior Citizens vs Churn (%)')
plt.xlabel('Senior Citizen (0 = No, 1 = Yes)')
plt.ylabel('Percentage')
plt.legend(title='Churn')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

```python
# List of columns to plot
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set figure size and layout
n_cols = 3
n_rows = (len(cols) + n_cols - 1) // n_cols  # Ceiling division
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 12))
axes = axes.flatten()

# Plot countplot for each column
for i, col in enumerate(cols):
    #sns.countplot(x=col, data=dataframe, ax=axes[i], palette='Set2')

    #sns.countplot(x=col, data=dataframe, ax=axes[i], hue=col, palette='Set2', legend=False,hue='Churn')
    sns.countplot(x=col, data=dataframe, ax=axes[i], hue='Churn', palette='Set2')


    #sns.countplot(x=col, data=dataframe, ax=axes[i], palette='Set2')

    
    axes[i].set_title(f"Count of {col}")
    axes[i].tick_params(axis='x', rotation=0)

# Hide unused subplots if any
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# This set of count plots shows the relationship between various services (like Phone, Internet, Security, Streaming) and customer churn. 
# Across most services, customers without additional services (like online security, backup, tech support) tend to churn more often. 
# Fiber optic internet and lack of protection services appear to correlate with higher churn rates.
```
